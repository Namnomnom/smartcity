```python
# Imports

import torch
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import requests
from io import BytesIO
import time
from datetime import datetime
from models import M_SFANet_UCF_QNRF
```

```python
# Modell laden
model = M_SFANet_UCF_QNRF.Model()
model.load_state_dict(torch.load("./Paper's_weights_UCF_QNRF/best_M-SFANet_UCF_QNRF.pth", map_location=torch.device('cpu')))
model.eval()
```

```python
# Aktuelles Webcam-Bild laden
img_url = "https://webcams.goslar.de/marktplatz-gmg.jpg"
response = requests.get(img_url)

if response.status_code == 200:
    img = Image.open(BytesIO(response.content)).convert('RGB')
    img.save("aktuelles_webcam_bild.jpg")
else:
    raise Exception("Bild konnte nicht geladen werden.")
```

```python
# Transformieren
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Größe anpassen (Mehrfaches von 16)
w, h = img.size
w = (w // 16) * 16
h = (h // 16) * 16
img = img.resize((w, h), Image.Resampling.LANCZOS)

# Tensor erzeugen
input_tensor = transform(img).unsqueeze(0)
```

```python
# Inferenz & Visualisierung

def monitor_webcam(interval_minutes=5, cycles=10):
    for i in range(cycles):
        print(f"\n⏱️ Durchlauf {i+1} von {cycles} ({datetime.now().strftime('%H:%M:%S')})")

        # 1. Bild laden
        img_url = "https://webcams.goslar.de/marktplatz-gmg.jpg"
        response = requests.get(img_url)
        if response.status_code != 200:
            print("❌ Fehler beim Laden des Bildes.")
            continue

        img = Image.open(BytesIO(response.content)).convert('RGB')
        timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
        filename = f"webcam_{timestamp}.jpg"
        img.save(filename)

        # 2. Bild vorbereiten
        w, h = img.size
        w = (w // 16) * 16
        h = (h // 16) * 16
        img = img.resize((w, h), Image.Resampling.LANCZOS)

        transform = transforms.Compose([
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                 std=[0.229, 0.224, 0.225])
        ])
        input_tensor = transform(img).unsqueeze(0).to('cpu')

        # 3. Modell-Inferenz
        with torch.no_grad():
            output = model(input_tensor)
        count = output.sum().item()

        # 4. Heatmap erstellen
        density_map = output.squeeze().cpu().numpy()
        density_map -= density_map.min()
        density_map /= (density_map.max() + 1e-9)
        density_map *= 255
        density_map = density_map.astype('uint8')

        # 5. Anzeige & Speicherung
        fig, axs = plt.subplots(1, 2, figsize=(10, 4))
        axs[0].imshow(img)
        axs[0].axis('off')
        axs[0].set_title("Webcam")

        axs[1].imshow(density_map, cmap='jet')
        axs[1].axis('off')
        axs[1].set_title(f"Geschätzte Personenanzahl: {count:.2f}")

        plt.tight_layout()
        # heatmap_file = f"heatmap_{timestamp}.png"
        # plt.savefig(heatmap_file)
        plt.show()
        plt.close()
        # print(f"✅ Heatmap gespeichert als: {heatmap_file}")

        # 6. Pause
        time.sleep(interval_minutes * 60)
```

```python
from IPython.display import clear_output

while True:
    monitor_webcam(interval_minutes=3, cycles=1)
    clear_output(wait=True)
```
